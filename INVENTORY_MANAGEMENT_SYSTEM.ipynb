```python
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *
import csv

class InventoryManagementSystem(QMainWindow):
    def __init__(self):
        super().__init__()
        
        # set up the main window
        self.setWindowTitle("Inventory Management System")
        self.setGeometry(100, 100, 800, 600)
        
        # set up the UI elements
        self.item_id_label = QLabel("Item ID:", self)
        self.item_id_label.move(50, 50)
        self.item_id_input = QLineEdit(self)
        self.item_id_input.move(150, 50)
        
        self.item_name_label = QLabel("Item Name:", self)
        self.item_name_label.move(50, 100)
        self.item_name_input = QLineEdit(self)
        self.item_name_input.move(150, 100)
        
        self.quantity_label = QLabel("Quantity:", self)
        self.quantity_label.move(50, 150)
        self.quantity_input = QSpinBox(self)
        self.quantity_input.move(150, 150)
        
        self.price_label = QLabel("Price:", self)
        self.price_label.move(50, 200)
        self.price_input = QDoubleSpinBox(self)
        self.price_input.move(150, 200)
        
        self.add_button = QPushButton("Add Item", self)
        self.add_button.move(50, 250)
        self.add_button.clicked.connect(self.add_item)
        
        self.add_button = QPushButton(" edit item" ,self)
        self.add_button.move(50, 300)
        self.add_button.clicked.connect(self.edit_item)
        
        
        self.export_button = QPushButton(" search item ", self)
        self.export_button.move(50, 350)
        self.export_button.clicked.connect(self.search_item)
        
        
        self.export_button = QPushButton("Generate report in database ", self)
        self.export_button.move(50, 400)
        self.export_button.clicked.connect(self.export_inventory_to_csv)
        
        self.statusBar().showMessage("Ready.")
        
        # set up the inventory list
        self.inventory = []
        
    def add_item(self):
        """
        Adds a new item to the inventory list based on the user input.
        """
        item_id = self.item_id_input.text()
        item_name = self.item_name_input.text()
        quantity = self.quantity_input.value()
        price = self.price_input.value()
        
        item = {'Item ID': item_id, 'Item Name': item_name, 'Quantity': quantity, 'Price': price}
        self.inventory.append(item)
        
        self.statusBar().showMessage("Item added successfully.")
        
        
    def edit_item(self):
        """
        Edits an existing item in the inventory list based on the user input.
        """
        item_id = self.item_id_input.text()
        item_name = self.item_name_input.text()
        quantity = self.quantity_input.value()
        price = self.price_input.value()
        
        for i, item in enumerate(self.inventory):
            if item['Item ID'] == item_id:
                self.inventory[i] = {'Item ID': item_id, 'Item Name': item_name, 'Quantity': quantity, 'Price': price}
                self.statusBar().showMessage("Item edited successfully.")
                return
                
        self.statusBar().showMessage("Item not found.")
        
        
        
        
    def search_item(self):
        """
        Edits an existing item in the inventory list based on the user input.
        """
        item_id = self.item_id_input.text()
        item_name = self.item_name_input.text()
        quantity = self.quantity_input.value()
        price = self.price_input.value()
        
        for i, item in enumerate(self.inventory):
            if item['Item ID'] == item_id:
                self.inventory[i] = {'Item ID': item_id, 'Item Name': item_name, 'Quantity': quantity, 'Price': price}
                self.statusBar().showMessage("Item edited successfully.")
                return
                
        self.statusBar().showMessage("Item not found.")
        
    def export_inventory_to_csv(self):
       
        
        
        
        
        
        
        
        """
        Exports the inventory data to a CSV file.
        """
        with open('inventory.csv', mode='w', newline='') as csv_file:
            fieldnames = ['Item ID', 'Item Name', 'Quantity', 'Price']
            writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

            writer.writeheader()

            for item in self.inventory:
                writer.writerow(item)
                
        self.statusBar().showMessage("Inventory exported to CSV file.")
        
if __name__ == '__main__':
    app = QApplication([])
    ims = InventoryManagementSystem()
    ims.show()
    app.exec_()
```
